```python
import psycopg2
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

conn = psycopg2.connect(
    user="your_username", 
    host="localhost", 
    password="your_password", 
    database="your_database", 
    port = 5432)

df = pd.read_sql("SELECT * FROM superstore", conn)
df = df.dropna()
for col in ('order_date', 'ship_date'):
    df[col] = pd.to_datetime(df[col], dayfirst=True)
    
df["year"] = df["order_date"].dt.year
df["month"] = df["order_date"].dt.month
```

```python
# monthly revenue trend
monthly_revenue_trend = df.groupby(["year", "month"], as_index=False)["sales"].sum()
# print(monthly_revenue_trend)
```

```python
# profit by category
category_profit = df.groupby(["category"])["sales"].sum()
# print(category_profit)
```

```python
# top customers visualization
top_customers = df.groupby(["customer_name"], as_index=False)["sales"].sum().sort_values(by="sales", ascending=False).head(10)
# print(top_customers)

x = top_customers["customer_name"]
y = top_customers["sales"]

plt.figure(figsize=(5, 5))
plt.bar(x, y)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# revenue trend chart
revenue_trend = df.groupby(["year"], as_index=False)["sales"].sum()

# matplotlib
x = revenue_trend["year"]
y = revenue_trend["sales"]
plt.plot(x, y)
plt.xticks(x)
plt.title("Revenue trend by year")
plt.show()

#seaborn
sns.lineplot(x = 'year', y = 'sales', data = revenue_trend);
plt.xticks(revenue_trend['year']);
plt.title("Revenue trend by year");
```

```python
# category comparison bar chart
category_comparison = df.groupby(["category"], as_index= False)["sales"].sum()

# matplotlib
x = category_comparison["category"]
y = category_comparison["sales"]
plt.bar(x, y)
plt.xticks(rotation= 45, ha = 'right')
plt.title("Category comparison")
plt.show()

#seaborn
sns.barplot(data= category_comparison, x = 'category', y = 'sales')
plt.xticks(rotation= 45, ha = 'right');
plt.title("Category comparison");
```

```python
# top customer chart
# top_customers = df.groupby(["customer_name"], as_index=False)["sales"].sum().sort_values(by= "sales", ascending = False).head(10)
top_customers = top_customers.sort_values("sales")

# matplotlib
x = top_customers["customer_name"]
y = top_customers["sales"]
plt.figure(figsize=(10, 6))
plt.barh(top_customers["customer_name"], top_customers["sales"])
plt.xlabel("sales")
plt.ylabel("customers")
plt.xticks(rotation = 45, ha = "right")
plt.title("Top customers chart")
plt.tight_layout()
plt.show()

# #seaborn
sns.barplot(data= top_customers, x = 'sales', y = 'customer_name');
plt.xticks(rotation = 45, ha = "right");
plt.title("Top customers chart");

```
